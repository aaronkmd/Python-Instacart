## 3. Create a Histogram of "order_hour_of_day"

### You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.

```python
# Creating a histogram for the "order_hour_of_day" column
hist_order_hour = df_visuals['order_hour_of_day'].plot.hist(bins=24, color='green', edgecolor='black')
```

```python
# Re-create the histogram with a title and labels
hist_order_hour = df_visuals['order_hour_of_day'].plot.hist(bins=24, color='green', edgecolor='black')
# Add a title
plt.title('Distribution of Orders by Hour of Day')
# Add a label for the x-axis
plt.xlabel('Hour of Day')  
# Add a label for the y-axis
plt.ylabel('Number of Orders') 
# Show the finished visualization
plt.show()
```

## 6. Create Line Chart to Explore Connections Between Age and Number of Dependents

### Now that you have information about customers, you need to conduct some exploratory analysis of customer demographics to inform the targeted marketing campaigns. ## 7. Create a Scatterplot of Age vs Income

### You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function. 

```python
# Reduce the previously created "small" sample to only the "age" and "income" columns
age_income = small[['age', 'income']]
```

```python
# Create the scatterplot visualizing the relationship between age and income
scatter_age_income = sns.scatterplot(x = 'age', y = 'income', data = age_income)
```

```python
# Re-create the scatterplot with a title and labels
scatter_age_income = sns.scatterplot(x = 'age', y = 'income', data = age_income)
# Add a title
plt.title('Age vs Income Scatterplot')
# Label the x-axis
plt.xlabel('Age')
# Label the y-axis
plt.ylabel('Income')
# Show the finished visualization
plt.show()
```

### This scatterplot shows a dense cluster around 200,000 and under across all age groups. There is also another cluster, not as dense, from age 40-80 with income between 200,000 and 300,000. There also appears to be quite a few outliers in the very high income range. These insights might help Instacart improve the targeting of their marketing campaigns, potentially pushing higher priced items to the demographic age 40-80 with >200,000 income.
